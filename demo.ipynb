```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n=50
N=1000

x = np.linspace(-3, 3, num=n)

X = np.linspace(-3, 3, num=N)
pix = np.pi*x
y = np.sin(pix)/pix + 0.1*x + 0.05*np.random.randn(n)

p = np.ones((29, n))
P = np.ones((29, N))
for j in range(1,15):
    p[2*j-1] = np.sin(x*j/2)
    p[2*j] = np.cos(x*j/2)
    P[2*j-1] = np.sin(X*j/2)
    P[2*j] = np.cos(X*j/2)

    
t = np.linalg.inv(np.dot(p, p.T)).dot(p).dot(y)
F = np.dot(P.T,t)
e = np.diag(np.append(np.ones(10),np.zeros(19)))
p1 = np.dot(p.T, e)
t2 = np.linalg.pinv(np.dot(p1.T, p1)).dot(p).dot(y)
F2 = np.dot(P.T, t2)


plt.scatter(x, y)
plt.plot(X, F, 'r')
plt.plot(X, F2, 'g')
plt.show()
```
