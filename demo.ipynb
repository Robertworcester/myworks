```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
n=50
N=1000
#x = np.array(list(-3+i*6/n for i in range(n)))
x = np.linspace(-3, 3, num=n)
#X = np.array(list(-3 + i*6/N for i in range(N)))
X = np.linspace(-3, 3, num=N)
pix = np.pi*x
y = np.sin(pix)/pix + 0.1*x + 0.05*np.random.randn(n)

p = np.ones(n)
P = np.ones(N)
for j in range(1,15):
    p = np.append(p,np.sin(x*j/2))
    p = np.append(p,np.cos(x*j/2))
    P = np.append(P,np.sin(X*j/2))
    P = np.append(P,np.cos(X*j/2))

p=p.reshape(29,n)
P=P.reshape(29,N)

#P = np.mat(P)
t = np.dot(np.dot(np.linalg.inv(np.dot(p, p.T)), p), y)
F = np.dot(P.T,t)


plt.scatter(x, y)
plt.plot(X, F, 'r')
plt.show()
```
